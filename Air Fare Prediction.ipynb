#  Working On Dept Time

```python
df["DEP_HOUR"]=pd.to_datetime(df.Dep_Time).dt.hour
df["DEP_MIN"]=pd.to_datetime(df.Dep_Time).dt.minute
df.drop(["Dep_Time"],axis=1)
```

#  Working on Arrival

```python
df["ARRIVAL_HOUR"]=pd.to_datetime(df.Arrival_Time).dt.hour
df["ARRIVAL_MIN"]=pd.to_datetime(df.Arrival_Time).dt.minute
df.drop(["Arrival_Time"],axis=1)
```

#  Working on Date of Journey

```python
df.info()
```

```python
df["journey_day"]=pd.to_datetime(df["Date_of_Journey"], format="%d/%m/%Y").dt.day
df["journey_month"]=pd.to_datetime(df["Date_of_Journey"], format="%d/%m/%Y").dt.month
df.drop(["Date_of_Journey"],axis=1)
```

#  working on total stops

```python
df["Total_Stops"].unique()
```

```python
order_Label={'4 stops':4,'3 stops':3,'2 stops':2,'1 stop':1,'non-stop':0 }
```

```python
df["Total_Stops"]=df["Total_Stops"].map(order_Label)
```

```python
df.head()
```

# Working on Airline

```python
airlines=df["Airline"].unique()

airlines
```

```python
for cat_var in airlines :
     df["Airline"+cat_var]=np.where(df["Airline"]==cat_var,1,0)
```

```python
df1=df.drop(columns=["Airline","Date_of_Journey","Arrival_Time","Dep_Time"],axis=1)
```

```python
df1.head()
```

#  Working on Source

```python
df1["Source"].unique()
```

```python
for cat_var in ['Banglore', 'Kolkata', 'Delhi', 'Chennai', 'Mumbai']:
    df1["source"+ cat_var]=np.where(df["Source"]==cat_var,1,0)
```

```python
df1.head()
```

#  working on destination

```python
df["Destination"].unique()
```

```python
for cat_var in ['New Delhi', 'Banglore', 'Cochin', 'Kolkata', 'Delhi', 'Hyderabad']:
    df1["Destination"+ cat_var]=np.where(df1["Destination"]==cat_var,1,0)
```

```python
df1.head()
```

```python
df2=df1.drop(columns=["Source","Destination","Additional_Info","Route","Duration"])
```

```python
df2.head()
```

```python
x=df2.drop(["Price"],axis=1)
y=df2["Price"]
print("shape of x is : ",x.shape)
print("shape of y is : ",y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=51)
print("shape of x_train is : ",x_train.shape)
print("shape of x_test is : ",x_test.shape)
print("shape of y_train is : ",y_train.shape)
print("shape of y_test is : ",y_test.shape)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x_train)
sc.transform(x_train)
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
```

#  Machine Learning Model Building Algorithms

#  Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
```

```python
lr=LinearRegression()
lr.fit(x_train_sc,y_train)
lr.score(x_test_sc,y_test)
```

```python
y_pred=lr.predict(x_test_sc)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(rmse)
```

#  Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtr.score(x_test,y_test)
```

```python
y_pred=dtr.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(rmse)
```

# Random Forest  Regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
rfr.score(x_test,y_test)
```

```python
y_pred=rfr.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(rmse)
```

#  XGBoost

```python
import xgboost as xgb
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb.fit(x_train,y_train)
xgb.score(x_test,y_test)
```

```python
y_pred=xgb.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(rmse)
```
