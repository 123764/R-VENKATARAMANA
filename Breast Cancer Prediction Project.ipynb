#  Split data

```python
from sklearn.model_selection import train_test_split
```

```python
x=df1.drop(["target"],axis=1)
x.shape
```

```python
y=df1["target"]
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=51)
print("shape of x_train: ",x_train.shape)
print("shape of x_test: ",x_test.shape)
print("shape of y_train: ",y_train.shape)
print("shape of y_test: ",y_test.shape)
```

#  Model Building

##  Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x_train)
sc.transform(x_train)
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

#  Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train_sc,y_train)
xlr1=lr.predict(x_test_sc)
accuracy_score(y_test,xlr1)
```

#  Support Vector Machine

```python
from sklearn.svm import SVC
svm_cls=SVC()
svm_cls.fit(x_train_sc,y_train)
y_svmm=svm_cls.predict(x_test_sc)
accuracy_score(y_test,y_svmm)
```

```python
svm_cls.score(x_test_sc,y_test)
```

#  Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
dtc_class=DecisionTreeClassifier(criterion='entropy',random_state=51)
dtc_class.fit(x_train,y_train)
y_pred_dtc=dtc_class.predict(x_test)
accuracy_score(y_test,y_pred_dtc)
```

```python
dtc_class.score(x_test,y_test)
```

#  Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier=RandomForestClassifier()
rf_classifier.fit(x_train,y_train)
rfcc=rf_classifier.predict(x_test)
accuracy_score(y_test,rfcc)
```

```python
rf_classifier.score(x_test,y_test)
```

#  k Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knc1_clas=KNeighborsClassifier()
knc1_clas.fit(x_train_sc,y_train)
kncc1=knc1_clas.predict(x_test_sc)
accuracy_score(y_test,kncc1)
```

# Naive Bayes Classification

```python
from sklearn.naive_bayes import GaussianNB
nbc_class=GaussianNB()
nbc_class.fit(x_train,y_train)
y_pred_nbc=nbc_class.predict(x_test)
accuracy_score(y_test,y_pred_nbc)
```

 # AdaBoost Classification 

```python
from sklearn.ensemble import AdaBoostClassifier
ada_classify=AdaBoostClassifier()
ada_classify.fit(x_train,y_train)
y_pred_ada=ada_classify.predict(x_test)
accuracy_score(y_pred_ada,y_test)
```

#  XGBoost Classification

```python
import xgboost as xgb
```

```python
from xgboost import XGBClassifier
xg_classifier=XGBClassifier()
xg_classifier.fit(x_train,y_train)
y_xg_classifier=xg_classifier.predict(x_test)
accuracy_score(y_xg_classifier,y_test)
```

# XGBoost Hyper  Parameters Tuning Randomized Search

```python
params={"learning_rate"    : [0.05,0.10,0.15,0.20,0.25,0.30],
       "max_depth"         : [3,4,5,6,8,10,12,15],
       "min_child_weight"  : [1,3,5,7] ,
       "gamma"             : [0.0,0.1,0.2,0.3,0.4],
       "colsample_bytree"  : [0.3,0.4,0.5,0.6],
       "n_estimators"      : [100,300,500],
       "objective"         : ["reg:linear"]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rsv=RandomizedSearchCV(xg_classifier,param_distributions=params,scoring="roc_auc",n_jobs=-1,verbose=True)
rsv.fit(x_train,y_train)
```

```python
rsv.best_params_
```

```python
rsv.best_estimator_
```

# Train Classifier with Best Estimators

```python
from xgboost import XGBClassifier
xgb_class = XGBClassifier(colsample_bytree=0.3, gamma=0.0, learning_rate=0.2, max_depth=4,
              min_child_weight=3, n_estimators=500,)
xgb_class.fit(x_train,y_train)
y_xgb_class=xgb_class.predict(x_test)
accuracy_score(y_xgb_class,y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_xgb_class)
plt.title("confusion matrix")
sns.heatmap(cm, annot=True)
plt.show()
```

#  Cross Validation of ML model

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_validation=cross_val_score(xgb_class, x_train , y_train , cv=10)
print(cross_validation)
print("mean:",cross_validation.mean)
```
